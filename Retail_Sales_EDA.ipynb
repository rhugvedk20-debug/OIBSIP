```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()

monthly_sales.plot(figsize=(12,5), marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
quarterly_sales = df.groupby(df['Order Date'].dt.to_period('Q'))['Sales'].sum()

quarterly_sales.plot(figsize=(10,5), marker='o', color='green')
plt.title("Quarterly Sales Trend")
plt.xlabel("Quarter")
plt.ylabel("Sales")
plt.show()
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='bar', figsize=(12,5))
plt.title("Top 10 Best-Selling Products")
plt.xlabel("Product Name")
plt.ylabel("Total Sales")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

category_sales.plot(kind='bar', figsize=(8,5))
plt.title("Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.show()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Correlation matrix
corr = numeric_df.corr()

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
segment_sales = df.groupby('Segment')['Sales'].sum().sort_values(ascending=False)

segment_sales.plot(kind='bar', figsize=(7,5))
plt.title("Sales by Customer Segment")
plt.xlabel("Segment")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Discount', y='Profit')
plt.title("Impact of Discount on Profit")
plt.show()
```

```python
Conclusion:

1. Sales show seasonal trends with peaks in specific months and quarters.
2. Technology / top products contribute the highest revenue.
3. Certain customer segments generate more sales than others.
4. Discounts negatively impact profitability in many cases.

Business Recommendations:
- Focus marketing on high-performing product categories.
- Optimize discount strategy to protect profit margins.
- Target high-value customer segments for growth.
```
